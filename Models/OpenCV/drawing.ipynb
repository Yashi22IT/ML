```python
import os
import cv2
```

```python
img = cv2.imread(os.path.join('D:\ML\Data','Wboard.jpg'))
```

```python
img.shape #height,width
```

```python
#line
cv2.line(img, (100,150), (250,300), (0,255,0), 3)
#(starting pt(x,y)), (end pt), (color), thickness
```

```python
#rectangle
cv2.rectangle(img, (50,150), (150,300), (0,0,255), 3)
cv2.rectangle(img, (50,100), (75,125), (0,0,255), -1)
```

```python
#circle
cv2.circle(img, (180,180),15,(255,0,0),3)
#center, radius, color, thickness
```

```python
#text
cv2.putText(img, 'Hey, you!',(250,200), cv2.FONT_HERSHEY_SIMPLEX,1, (255,0,255),2)
#string, starting pt, font, font size, color, thickness
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
```
